```python
# First removing the columns I do not need  
new_cars = cars.drop(['Transaction ID', 'VEHICLE ID', 'VEHICLETYPE'], axis = 1)
new_cars.set_index('SOLD DATE')
```

# Feature Engineering (Deriving new features) for categorical data 

I used two main encoding techniques for catergorial data 
 - Ordinal Encoding 
 - One-hot Encoding 

```python
#Encode DriveTrain with one-hot encoding
# initiate new columns for Allwheel and frontwheel drive 
new_cars['AllWheel'] = new_cars['CONDITION GRADE']
new_cars['FrontWheel'] = new_cars['CONDITION GRADE']

# onehot the DriveTrain to AllWheels/FrontWheels

for i in range(len(new_cars['DRIVETRAIN'])): #array(['4x2', '4x4'], dtype=object)
  if new_cars['DRIVETRAIN'][i] == '4x2':
    new_cars['AllWheel'][i] = 0
    new_cars['FrontWheel'][i] = 1
  else:
    new_cars['AllWheel'][i] = 1
    new_cars['FrontWheel'][i] = 0

```

```python
#Encode petrol and diesel fuel with one-hot encoding 
new_cars['Petrol'] = new_cars['CONDITION GRADE']
new_cars['Diesel'] = new_cars['CONDITION GRADE']
# now onehot code two main fuel 'Petrol', 'Diesel'
for i in range(len(new_cars['FUEL'])): #array(['Petrol', 'Diesel'], dtype=object)
  if new_cars['FUEL'][i] == 'Petrol':
    new_cars['Petrol'][i] = 1
    new_cars['Diesel'][i] = 0
  else:
    new_cars['Petrol'][i] = 0
    new_cars['Diesel'][i] = 1

```

```python
# Ordinal encoding for Conutry Name sorted by mean price of MSRP
new_cars['Country'] = new_cars['CONDITION GRADE']

for i in range(len(new_cars['COUNTRYOFORIGIN_ADJ'])):
  if new_cars['COUNTRYOFORIGIN_ADJ'][i] == 'India':
    new_cars['Country'][i] = 1
  elif new_cars['COUNTRYOFORIGIN_ADJ'][i] == 'Mexico':
    new_cars['Country'][i] = 2
  elif new_cars['COUNTRYOFORIGIN_ADJ'][i] == 'Korea':
    new_cars['Country'][i] = 3
  elif new_cars['COUNTRYOFORIGIN_ADJ'][i] == 'Great Britain':
    new_cars['Country'][i] = 4
  elif new_cars['COUNTRYOFORIGIN_ADJ'][i] == 'Spain':
    new_cars['Country'][i] = 5
  elif new_cars['COUNTRYOFORIGIN_ADJ'][i] == 'Japan':
    new_cars['Country'][i] = 6
  elif new_cars['COUNTRYOFORIGIN_ADJ'][i] == 'Czech Republic':
    new_cars['Country'][i] = 7
  elif new_cars['COUNTRYOFORIGIN_ADJ'][i] == 'Canada':
    new_cars['Country'][i] = 8
  elif new_cars['COUNTRYOFORIGIN_ADJ'][i] == 'Thailand':
    new_cars['Country'][i] = 9
  elif new_cars['COUNTRYOFORIGIN_ADJ'][i] == 'Germany':
    new_cars['Country'][i] = 10
  elif new_cars['COUNTRYOFORIGIN_ADJ'][i] == 'USA':
    new_cars['Country'][i] = 11
  elif new_cars['COUNTRYOFORIGIN_ADJ'][i] == 'South Africa':
    new_cars['Country'][i] = 12
  elif new_cars['COUNTRYOFORIGIN_ADJ'][i] == 'Slovakia':
    new_cars['Country'][i] = 13
  elif new_cars['COUNTRYOFORIGIN_ADJ'][i] == 'Austria':
    new_cars['Country'][i] = 14
  
```

```python
# Encoding the Engine Data with one-hot encoding 
#initlizing the columns 
new_cars['ENGINE SIZE'] = new_cars['CONDITION GRADE'] 
new_cars['ENGINE FORM'] = new_cars['CONDITION GRADE'] 
new_cars['ENGINE Type'] = new_cars['CONDITION GRADE'] 
new_cars['ENGINE NUMBER'] = new_cars['CONDITION GRADE']
new_cars['ENGINE CLASS'] = new_cars['CONDITION GRADE'] 

# spliting the ENGINE column to 5 columns 
for i in range(len(new_cars['ENGINE'])): 
  val = new_cars['ENGINE'][i].split()
  if len(val) == 6:
    new_cars['ENGINE SIZE'][i] = val[0][:-1]
    new_cars['ENGINE FORM'][i] = val[1]
    new_cars['ENGINE Type'][i] = val[2]
    new_cars['ENGINE NUMBER'][i] = val[3][0]
    new_cars['ENGINE CLASS'][i] = val[4]
  elif len(val) == 5:
    new_cars['ENGINE SIZE'][i] = val[0][:-1]
    new_cars['ENGINE FORM'][i] = val[1]
    new_cars['ENGINE Type'][i] = val[2]
    new_cars['ENGINE NUMBER'][i] = val[3][0]
    new_cars['ENGINE CLASS'][i] = val[4]
  elif len(val) == 4:
    new_cars['ENGINE SIZE'][i] = val[0][:-1]
    new_cars['ENGINE FORM'][i] = val[1]
    new_cars['ENGINE Type'][i] = val[2]
    new_cars['ENGINE NUMBER'][i] = val[3][0]
    new_cars['ENGINE CLASS'][i] = 0
  else:
    new_cars['ENGINE SIZE'][i] = val[0][:-1]
    new_cars['ENGINE FORM'][i] = 0
    new_cars['ENGINE Type'][i] = 0
    new_cars['ENGINE NUMBER'][i] = 0
    new_cars['ENGINE CLASS'][i] = 0

```

```python
# Embedding the new 5 Engine Columns 
print(new_cars['ENGINE SIZE'].unique()) # [1.4 1.5 2.0 1.6 2.2 2.8 2.5 3.5 3.2 1.0 3.0 3.6 1.8 2.4 1.3 5.6 4.0 2.9 2.1 4.4 1.7 1.2 'Electri' '1.4' '2.0' '3.5' '2.2' '3.6' '2.5' '3.2' '1.6'
 #'2.8' '1.5' '3.0' '2.4' '1.0' '1.2' '4.4' '5.6' '1.3' '1.8' '4.0' '2.1' '2.9']
print(new_cars['ENGINE FORM'].unique()) # ['In-Line' 'V' 0] 
print(new_cars['ENGINE Type'].unique()) # ['Turbo' '4-cyl' '6-cyl' 'Supercharger' '8-cyl' '3-cyl' 0]
print(new_cars['ENGINE NUMBER'].unique())  # [4.0 'U' '4' '5' '3' '6' 'P' '8' 0]
print(new_cars['ENGINE CLASS'].unique()) # ['Premium' 'Diesel' 0 'Unleaded' 'Electric']
```

```python
#Encoding the Engine Form, Type with one-hot enciding
new_cars['In-Line'] = new_cars['CONDITION GRADE']
new_cars['V-Shape'] = new_cars['CONDITION GRADE']
for i in range(len(new_cars['ENGINE FORM'])):
  if new_cars['ENGINE FORM'][i] == 'In-Line':
    new_cars['In-Line'][i] = 1 
    new_cars['V-Shape'][i] = 0
  elif new_cars['ENGINE FORM'][i] == 'V':
    new_cars['In-Line'][i] = 0
    new_cars['V-Shape'][i] = 1
  else:
    new_cars['In-Line'][i] = 0
    new_cars['V-Shape'][i] = 0

# One-hot the Engine Type 
new_cars['Turbo'] = new_cars['CONDITION GRADE']
new_cars['Supercharger'] = new_cars['CONDITION GRADE']


for i in range(len(new_cars['ENGINE Type'])):
  if new_cars['ENGINE Type'][i] == 'Turbo':
    new_cars['Turbo'][i] = 1
    new_cars['Supercharger'][i] = 0
  elif new_cars['ENGINE Type'][i] == 'Supercharger':
    new_cars['Turbo'][i] = 0
    new_cars['Supercharger'][i] = 1
  else:
    new_cars['Turbo'][i] = 0
    new_cars['Supercharger'][i] = 0

new_cars['regular'] = new_cars['CONDITION GRADE']
new_cars['Premium'] = new_cars['CONDITION GRADE']
new_cars['Cylinders'] = new_cars['CONDITION GRADE']

for i in range(len(new_cars['ENGINE NUMBER'])):
  if (new_cars['ENGINE CLASS'][i] == 'Premium') or (new_cars['ENGINE NUMBER'][i] == 'P'):
    new_cars['regular'][i] = 0 
    new_cars['Premium'][i] = 1
    if (new_cars['ENGINE NUMBER'][i] == 'P'):
      new_cars['Cylinders'][i] = int(new_cars['ENGINE Type'][i][0])
    else:
      new_cars['Cylinders'][i] = int(new_cars['ENGINE NUMBER'][i])
  else: 
    new_cars['regular'][i] = 1
    new_cars['Premium'][i] = 0
    new_cars['Cylinders'][i] = int(new_cars['ENGINE NUMBER'][0])
```

```python
# Encoding BodyType, SegmentName, SubSegmentname 
print(new_cars['BODYTYPE'].unique())
print(new_cars['SEGMENTNAME'].unique())
print(new_cars['SUBSEGMENTNAME'].unique())


# one-hot the body type ['Sedan' 'SUV/Crossover']
new_cars['Sedan'] = new_cars['CONDITION GRADE']
new_cars['SUV'] = new_cars['CONDITION GRADE']

for i in range(len(new_cars['BODYTYPE'])):
  if new_cars['BODYTYPE'][i] == 'Sedan':
    new_cars['Sedan'][i] = 1
    new_cars['SUV'][i] = 0
  else:
    new_cars['Sedan'][i] = 0
    new_cars['SUV'][i] = 1


# Nomial encoding for the size of the car 

new_cars['CarSize'] = new_cars['CONDITION GRADE']
for i in range(len(new_cars['SEGMENTNAME'])):
  splited = new_cars['SEGMENTNAME'][i].split()
  if 'Light/Micro' in splited:
    new_cars['CarSize'][i] = 1
  elif 'Small' in splited:
    new_cars['CarSize'][i] = 2
  elif 'Medium' in splited:
    new_cars['CarSize'][i] = 3
  elif 'Large' in splited:
    new_cars['CarSize'][i] = 4
  else: #pickup
    new_cars['CarSize'][i] = 5 
    
#One-hot encoding of luxury, Economy, mind-range 

new_cars['Economic'] = new_cars['CONDITION GRADE']
new_cars['Mid-rang'] = new_cars['CONDITION GRADE']
new_cars['Luxury'] = new_cars['CONDITION GRADE']

for i in range(len(new_cars['SUBSEGMENTNAME'])):
  splited = new_cars['SUBSEGMENTNAME'][i].split()
  if 'Discount' in splited:
    new_cars['Economic'][i] = 1
    new_cars['Mid-rang'][i] = 0
    new_cars['Luxury'][i] = 0
  elif 'Luxury' in splited:
    new_cars['Economic'][i] = 0
    new_cars['Mid-rang'][i] = 0
    new_cars['Luxury'][i] = 1
  else:
    new_cars['Economic'][i] = 0
    new_cars['Mid-rang'][i] = 1
    new_cars['Luxury'][i] = 0

```

```python
#Nomial Encoding of the ModelName
print(new_cars['MODEL NAME'].unique())  
# sorted mode name by their average MSRP 
sorted_model_name = ['Accent', 'Venue', 'Astra', 'EcoSport', 'Elantra', 'Trax', 'Kona',
       'Sonata', 'Focus', 'Juke', 'Qashqai', 'Tucson', 'Captiva', 'X-TRAIL',
       'Equinox', 'Escape', 'i40', 'GLA-Class', 'A3', 'Trailblazer',
       'Santa Fe', 'X1', 'X2', 'Acadia', 'A-Class', 'Pathfinder', 'A4',
       'Endura', 'Everest', 'GLC-Class', '3 Series', 'C-Class', 'Patrol', 'X4',
       'X3', 'GLC', 'GLE-Class', 'GLE', 'Q7', '5 Series', 'X5', 'GLS-Class',
       'E-Class', 'X6', 'X7', 'S-Class', '7 Series', 'G-class']

# let's assign valude(their index) to each model name 
dic_model = {}
for i, val in enumerate(sorted_model_name):
  dic_model[val] = i

new_cars['Model'] = new_cars['CONDITION GRADE']

for i in range(len(new_cars['MODEL NAME'])):
  new_cars['Model'][i] = dic_model[new_cars['MODEL NAME'][i]]

```

```python
# Nomial Encoding of Brand Name  
# sorted brand name by their average MSRP 
sorted_brand_name = ['Holden', 'Hyundai', 'Nissan', 'Ford', 'Mercedes-Benz', 'BMW', 'Audi']

dic_brand = {}
for i, val in enumerate(sorted_brand_name):
  dic_brand[val] = i

new_cars['Brand'] = new_cars['CONDITION GRADE']

for i in range(len(new_cars['BRAND NAME'])):
  new_cars['Brand'][i] = dic_brand[new_cars['BRAND NAME'][i]]

# print(new_cars['BRAND NAME'].unique())
# a = new_cars.groupby(by = ['BRAND NAME']).mean()['MSRP']
# b= a.sort_values()
# c = b.index
# c
```
